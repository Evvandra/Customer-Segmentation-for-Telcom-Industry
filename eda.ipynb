```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python

# Assuming 'df' is your DataFrame
plt.figure(figsize=(12, 6))  # Set the figure size

# Group by 'OperatorName' and calculate the mean of 'SuccessCharging' for each operator
operator_success = df.groupby('OperatorName')['SuccessCharging'].sum().sort_values(ascending=False)

# Plotting the bar plot
sns.barplot(x=operator_success.index, y=operator_success.values, palette='viridis')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.xlabel('Operator Name')
plt.ylabel('Total Success Charging')
plt.title('Total Success Charging by Operator')
plt.tight_layout()

plt.show()
```

```python

# Assuming 'df' is your DataFrame
plt.figure(figsize=(12, 6))  # Set the figure size

# Group by 'OperatorName' and calculate the mean of 'SuccessCharging' for each operator
operator_success = df.groupby('CountryCode')['SuccessCharging'].sum().sort_values(ascending=False)

# Plotting the bar plot
sns.barplot(x=operator_success.index, y=operator_success.values, palette='viridis')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.xlabel('Country Code')
plt.ylabel('Total Success Charging')
plt.title('Total Success Charging by Country')
plt.tight_layout()

plt.show()
```
