```python
import pandas as pd

# Assuming df is your DataFrame with SubscriptionTime and RenewalTime columns
# Convert the time columns to datetime format
df['SubscriptionTime'] = pd.to_datetime(df['SubscriptionTime'])
df['RenewalTime'] = pd.to_datetime(df['RenewalTime'])

# Define the time categories
def categorize_time(time):
    if 5 <= time.hour < 12:
        return 'Morning'
    elif 12 <= time.hour < 17:
        return 'Afternoon'
    elif 17 <= time.hour < 21:
        return 'Evening'
    else:
        return 'Night'

# Apply the categorization to SubscriptionTime and RenewalTime
df['SubscriptionTimeCategory'] = df['SubscriptionTime'].apply(categorize_time)
df['RenewalTimeCategory'] = df['RenewalTime'].apply(categorize_time)

# Display the updated DataFrame
print(df[['SubscriptionTime', 'SubscriptionTimeCategory', 'RenewalTime', 'RenewalTimeCategory']])
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame with RenewalTimeCategory and SuccessCharging columns
# Group by RenewalTimeCategory and calculate the sum of SuccessCharging for each category
grouped_data = df.groupby('RenewalTimeCategory')['SuccessCharging'].sum().reset_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='RenewalTimeCategory', y='SuccessCharging', data=grouped_data, palette='viridis')
plt.title('Total Success Billing by Renewal Time Category')
plt.xlabel('Time')
plt.ylabel('Total Success Billing')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame with RenewalTimeCategory and SuccessCharging columns
# Group by RenewalTimeCategory and calculate the sum of SuccessCharging for each category
grouped_data = df.groupby('SubscriptionTimeCategory')['SuccessCharging'].sum().reset_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='SubscriptionTimeCategory', y='SuccessCharging', data=grouped_data, palette='viridis')
plt.title('Total Success Billing by Subscription Time Category')
plt.xlabel('Time')
plt.ylabel('Total Success Billing')
plt.show()
```

```python
# Assuming 'df' is your DataFrame
plt.figure(figsize=(12, 6))  # Set the figure size

# Convert 'OperatorName' values to lowercase and group by, then calculate the sum of 'SuccessCharging' for each operator
operator_success = df.groupby(df['OperatorName'].str.lower())['SuccessCharging'].sum().sort_values(ascending=False)

# Plotting the bar plot
sns.barplot(x=operator_success.index, y=operator_success.values, palette='viridis')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.xlabel('Operator Name', fontsize=16)
plt.ylabel('Total Success Charging', fontsize=16)
plt.title('Total Success Charging by Operator', fontsize=20)
plt.tight_layout()
plt.savefig('Success_operator.png')
plt.show()
```

```python
import pandas as pd

# Replace values containing "star" with "star"
df['PublisherName'] = df['PublisherName'].str.replace('.*star.*', 'star', regex=True)
df['PublisherName'] = df['PublisherName'].str.replace('.*lig.*', 'lig', regex=True)


# Display the updated DataFrame
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named df with your data
# If you don't have a DataFrame, you can create one using pandas.read_csv() or any other method

# Convert AdnetName and PublisherName columns to lowercase
df['AdnetName'] = df['AdnetName'].str.lower()
df['PublisherName'] = df['PublisherName'].str.lower()

# Filter rows with successful billing
success_df = df[df['SuccessCharging'] == 1]

# Group by PublisherName and calculate the total success billing
billing_counts = success_df['PublisherName'].value_counts()

# Plot the bar graph with more colorful and larger settings
plt.figure(figsize=(10, 6))  # Set the figure size
billing_counts.plot(kind='bar', color=['lightblue', 'lightgreen', 'lightcoral', 'lightskyblue', 'lightpink'])
plt.xlabel('Publisher')
plt.ylabel('Total Success Billing')
plt.title('Total Success Billing by Publisher')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines for better visualization
plt.tight_layout()
plt.show()
```
