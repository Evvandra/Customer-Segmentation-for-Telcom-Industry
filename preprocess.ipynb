```python
import pandas as pd
```

```python
df_old = pd.read_csv('subs.csv')
df_new = pd.read_csv('subscriptions_60000.csv', delimiter=';')
```

```python
df_old['subs_date'] = pd.to_datetime(df_old['subs_date'], format='%Y-%m-%d %H:%M:%S.%f', errors='coerce')
df_old['subs_date'] = df_old['subs_date'].dt.strftime('%d/%m/%Y %H:%M')

df_old
```

```python
df_old['renewal_date'] = pd.to_datetime(df_old['renewal_date'], format='%Y-%m-%d %H:%M:%S.%f', errors='coerce')
df_old['renewal_date'] = df_old['renewal_date'].dt.strftime('%d/%m/%Y %H:%M')

df_old
```

```python
df_new['subs_date'] = pd.to_datetime(df_new['subs_date'], format='%Y-%m-%d %H:%M:%S.%f %z', errors='coerce')
df_new['subs_date'] = df_new['subs_date'].dt.strftime('%d/%m/%Y %H:%M')

df_new
```

```python
df_new['renewal_date'] = pd.to_datetime(df_new['renewal_date'], format='%Y-%m-%d %H:%M:%S.%f %z', errors='coerce')
df_new['renewal_date'] = df_new['renewal_date'].dt.strftime('%d/%m/%Y %H:%M')

df_new
```

```python
df = pd.concat([df_old, df_new], ignore_index=True)
df
```

```python
df.sort_values(by='id', inplace=True)
```

```python
df.drop_duplicates()
```

```python
df = df[['id', 'country', 'operator', 'service', 'source', 'status', 'cycle', 'adnet',
         'subs_date', 'renewal_date', 'publisher', 'attempt_charging', 'success_billing']]
```

```python
df
```

```python
df[['Subsdate', 'Substime']] = df['subs_date'].str.split(expand=True)

# Converting 'Subsdate' column to datetime format
df['Subsdate'] = pd.to_datetime(df['Subsdate'], format='%d/%m/%Y')

# Converting 'Substime' column to time format
df['Substime'] = pd.to_datetime(df['Substime'], format='%H:%M').dt.time

# Drop the original 'subs_date' column if you no longer need it
df = df.drop(columns=['subs_date'])

# Display the modified DataFrame
df
```

```python
df[['Renewaldate', 'Renewaltime']] = df['renewal_date'].str.split(expand=True)

# Drop rows with '0001-01-01' in 'Renewaldate' column
df = df[df['Renewaldate'] != '0001-01-01']

# Converting 'Renewaldate' column to datetime format (assuming the format is '26/08/2023')
df['Renewaldate'] = pd.to_datetime(df['Renewaldate'], format='%d/%m/%Y')

# Converting 'Renewaltime' column to time format
df['Renewaltime'] = pd.to_datetime(df['Renewaltime'], errors='coerce').dt.time

# Drop the original 'renewal_date' column if you no longer need it
df = df.drop(columns=['renewal_date'])

# Display the modified DataFrame
df
```

```python
df.info()
```

```python
new_column_names = {
    'id': 'CustomerID',
    'country': 'CountryCode',
    'operator': 'OperatorName',
    'service': 'ServiceType',
    'source': 'SubscriptionChannel',
    'status': 'CustomerStatus',
    'cycle': 'SubscriptionCycle',
    'adnet': 'AdnetName',
    'publisher': 'PublisherName',
    'attempt_charging': 'ChargingAttempts',
    'success_billing': 'SuccessCharging',
    'Subsdate': 'SubscriptionDate',
    'Substime': 'SubscriptionTime',
    'Renewaldate': 'RenewalDate',
    'Renewaltime': 'RenewalTime'
}

# Assuming df is your DataFrame
df = df.rename(columns=new_column_names)
```

```python
df[['SubscriptionChannel', 'AdnetName', 'PublisherName']] = df[['SubscriptionChannel', 'AdnetName', 'PublisherName']].fillna('not found')

# Use a more reasonable maximum date that fits within the valid range
columns_to_drop = ['SubscriptionDate', 'RenewalDate', 'SubscriptionTime', 'RenewalTime']

# Fill missing values with the default datetime value
df = df.dropna(subset=columns_to_drop)
```
